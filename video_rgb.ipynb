```python
import cv2
import os
```

```python
src_dir = '/home/axel/Desktop/STAM/dataset/EGTEA_Gaze_plus/z_frames_rgb2/'
dst_dir = '/home/axel/Desktop/STAM/dataset/EGTEA_Gaze_plus/z_frames_rgb/'
```

```python
# for directory in os.listdir(src_dir):
#     os.mkdir(dst_dir + directory)
```

```python
for directory in os.listdir(src_dir):
    for videofile in os.listdir(src_dir + directory):
        #print(dst_dir + directory + '/' + videofile[:-4])
        
        os.mkdir(dst_dir + directory + '/' + videofile[:-4])
        vidcap = cv2.VideoCapture(src_dir + directory + '/' + videofile)
        
        success,image = vidcap.read()
        count = 1

        while success:
            cv2.imwrite(dst_dir + directory + '/' + videofile[:-4] + "/%06d.jpg" % count, image) # save frame as JPEG file      
            success,image = vidcap.read()
            #print('Read a new frame: ', success)
            count += 1
```
